```python
# Funções: gorjeta

def main():
  a = int(input("Digite o valor total da compra: "))
  b = int(input("Digite a porcentagem de gorjeta: "))

  calculo_gorjeta(a, b)

def calculo_gorjeta(valor, porcentagem):
  print((porcentagem/100) * valor)

main()
```
