```python
# Listas
minha_lista = [
    "João",
    "Maria",
    "José",
    "Pedro",
    "Fulano",
    "Ciclano",
    "Beltrano"
    ]

print(minha_lista)
print("A lista tem", len(minha_lista), "nomes até o momento.")
nome_escolha = input("Deseja adicionar algum nome?yes/no\n").strip()

if nome_escolha == "yes":
  nome_adicional = input("Digite o nome:\n").strip().capitalize()
  minha_lista.append(nome_adicional)
  print(minha_lista)
else:
  exit()
```

```python
# Conjuntos / sets
meu_set = {
    "João",
    "Maria",
    "José",
    "Pedro",
    "Fulano",
    "Ciclano",
    "Beltrano"
}

print(meu_set)
print("O conjunto tem", len(meu_set), "nomes até o momento.")
escolha_adicao = input("Deseja adicionar algum nome?yes/no\n").strip()

if escolha_adicao == "yes":
  nome_adicional = input("Digite o nome:\n").strip().capitalize()
  meu_set.add(nome_adicional)
  print(meu_set)

esoolha_remocao = input("Deseja remover algum nome?yes/no\n").strip()

if esoolha_remocao == "yes":
  nome_removido = input("Digite o nome:\n").strip().capitalize()
  meu_set.discard(nome_removido)
  print(meu_set)
```

```python
# Sets: união, intersecção e diferença
set1 = {
    "João",
    "Maria",
    "José",
    "Pedro",
    "Fulano",
    "Ciclano",
    "Beltrano"
}
set2 = {
    "Maria",
    "Carlos",
    "Arthur",
    "Fulano",
    "Marcos",
    "Joana"
}

print(set1.difference(set2))
print(set1.union(set2))
print(set1.intersection(set2))
print(set1.symmetric_difference(set2))
print(set1.isdisjoint(set2))
```

```python
# Comando for
colors = ["Red", "Green", "Blue", "Orange", "Black", "White", "Yellow"]

# Aqui, o underline é um número que vai printar as cores de acordo com a
# posição dentro do escopo (range) da length de colors
for _ in range(len(colors)):
  #print(_)
  print("Cor:" , colors[_] , "/ posição:", _)

# Já aqui, o for faz a iteração com o color assumindo cada valor individual
# das cores
for color in colors:
    print(color)
```

```python
# Classes - exemplo do site

class Pais: # definição do modelo

    def __init__(self, nome, continente, capital, idioma): # função construtora da classe. Responsável por definir as características do tipo de dado (no nosso caso, `Pais`)
        self.nome = nome
        self.continente = continente
        self.capital = capital
        self.idioma = idioma

    def localizacao(self): # criação de um método. Define determinado comportamento ou ação que um objeto pode realizar
        if self.continente:
            print(f'{self.nome} está localizado no {self.continente}')
        else:
            print('Continente não indicado')


    def __str__(self):
      return f'{self.nome} está localizado no continente {self.continente} e tem como capital {self.capital}. O idioma oficial é o {self.idioma}.'



def main(): #
    paises = Pais('Africa do Sul', 'África', 'Joanesburgo', 'Inglês') # criação de um objeto do tipo país que possui os atributos nome, continente, capital e idioma
    print(paises)
    print(f'{paises.nome}, {paises.continente}, {paises.capital},  {paises.idioma}')
    lugar = paises.localizacao()



if __name__ == "__main__":
    main()
```

```python
# Método __init__()

class Pessoa:
  # O __init__() serve para assinalar os valores às propriedades dos objetos,
  # ou realizar operações que são necessárias quando o objeto está sendo criado
  def __init__(self, nome, idade, cidade, email):
    self.nome = nome
    self.idade = idade
    self.cidade = cidade
    self.email = email

pessoa1 = Pessoa("Leonardo", 20, "Jundiaí", "abc@abc.com")

print(pessoa1.nome)
print(pessoa1.idade)
print(pessoa1.cidade)
print(pessoa1.email)
```

```python
# Self e métodos - meio exagerado, mas é possível chamar self sobre self
class Pessoa:
  def __init__(self, nome):
    self.nome = nome

  def msg_ola(self):
    return "Olá " + self.nome

  def msg_completa(self):
    ola = self.msg_ola()
    print(ola + "! Seja bem vindo")

pessoa1 = Pessoa("Leonardo")
pessoa1.msg_completa()
```

```python
# Class

class Calculadora:
  def multiplicacao(self, x, y):
    return x * y

  def divisao(self, x, y):
    return x / y

  def soma(self, x, y):
    return x + y

  def subtracao(self, x, y):
    return x - y

cal = Calculadora()
print(cal.multiplicacao(10, 5))
print(cal.divisao(10, 5))
print(cal.soma(10, 5))
print(cal.subtracao(10, 5))
```

```python
# __str__()

class Pessoa:
  def __init__(self, nome, cidade):
    self.nome = nome
    self.cidade = cidade

  def __str__(self) -> str:
    return self.nome + " mora em " + self.cidade

pessoa1 = Pessoa("Leonardo", "Franca")
print(pessoa1)
```
