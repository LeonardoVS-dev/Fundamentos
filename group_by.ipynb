Aqui, vou utilizar o arquivo de votaĂ§ĂŁo das eleiĂ§Ă”es de 2024 de Roraima para trabalhar com o group_by

```python
import pandas as pd
# ImportaĂ§ĂŁo do arquivo
#dados_eleitorais = pd.read_csv("/content/votacao_candidato_munzona_2024_RR.csv", encoding="latin-1", sep=";")
#dados_eleitorais = pd.read_csv("/content/votacao_candidato_munzona_2024_RR LIMPADO.csv", encoding="latin-1", sep=";")
dados_eleitorais = pd.read_csv("/content/votacao_candidato_munzona_2024_RR LIMPADO_CORRIGIDO.csv", encoding="latin-1", sep=";")
```

```python
# Aqui, crio um dataframe com base nos dados eleitorais importados
dataframe_dados_eleitorais = pd.DataFrame(dados_eleitorais)

# Aqui vai o primeiro group_by: ele serve para agrupar o nĂșmero do partido
# e o nĂșmero de votos nominais por cada partido, ou seja,
# traz a soma da contagem de votos por part.
group_numVotos_numPart = dataframe_dados_eleitorais['QT_VOTOS_NOMINAIS'].groupby(dataframe_dados_eleitorais['NR_PARTIDO'])

# No caso, utilizei a funĂ§ĂŁo sum() para trazer a soma de todos os votos que
# cada partido recebeu
print(group_numVotos_numPart.sum())
```

```python
# Aqui, crio um dataframe com base nos dados eleitorais importados
dataframe_dados_eleitorais = pd.DataFrame(dados_eleitorais)

# Aqui vai o segundo group_by: ele serve para agrupar o nĂșmero do partido,
# o nome do partido e o nĂșmero de votos nominais por cada partido, ou seja,
# traz a soma da contagem de votos por part.
group_numVotos_numPart = dataframe_dados_eleitorais['QT_VOTOS_NOMINAIS'].groupby([dataframe_dados_eleitorais['NR_PARTIDO'], dataframe_dados_eleitorais['NM_PARTIDO']])
print(group_numVotos_numPart.sum())
```

```python
dataframe_dados_eleitorais = pd.DataFrame(dados_eleitorais)

# Aqui, ele faz a contagem de quantos vereadores (cĂłdigo 13) e
# quantos prefeitos estiveram na disputa
print(dataframe_dados_eleitorais["CD_CARGO"].value_counts())
```

```python
dataframe_dados_eleitorais = pd.DataFrame(dados_eleitorais)

# Aqui, eu agrupo o nĂșmero de votos nominais com o cĂłdigo do cargo.
group_numVotos_cdCargo = dataframe_dados_eleitorais['QT_VOTOS_NOMINAIS'].groupby(dataframe_dados_eleitorais['CD_CARGO'])

# Assim, sei qual foi o nĂșmero mĂ©dio de votos de cada prefeito e
# de cada vereador atravĂ©s do mean()
print(group_numVotos_cdCargo.mean())

# JĂĄ aqui ele retorna o nĂșmero max de votos que um vereador ou prefeito recebeu
print(group_numVotos_cdCargo.max())

# Aqui Ă© o mĂ­nimo
print(group_numVotos_cdCargo.min())
```

```python
# Group_by com vĂĄrios Ă­ndices
dataframe_dados_eleitorais = pd.DataFrame(dados_eleitorais)

#group_multiplo = dataframe_dados_eleitorais.groupby(["NM_MUNICIPIO", "NM_PARTIDO"]).head(30)

dataframe_dados_eleitorais.groupby(["NM_PARTIDO", "CD_CARGO"]).sum()
```

```python
dataframe_dados_eleitorais = pd.DataFrame(dados_eleitorais)

# Aqui, o Ă­ndice do group_by Ă© o nome de cada municĂ­pio. Com a funĂ§ĂŁo agg,
# utilizamos o mĂ©todo size() para contar o nĂșmero total de candidatos nos dois
# cargos em disputa (prefeito e vereador). AlĂ©m disso, ele traz o nĂșmero mĂ©dio
# de votos de cada candidato pelo mĂ©todo mean(). Ă interessante observar que Boa Vista, a capital
# do estado, Ă© a cidade que mais tem candidatos
dataframe_dados_eleitorais.groupby("NM_MUNICIPIO").agg({"CD_CARGO":"size", "QT_VOTOS_NOMINAIS":"mean"})
```

```python
# Filtragem pelo nome da cidade
dataframe_boavista = dataframe_dados_eleitorais[dataframe_dados_eleitorais["NM_MUNICIPIO"] == "BOA VISTA"]
# NĂșmero de candidatos por partido em Boa Vista
dataframe_boavista.groupby("NM_PARTIDO").size()
```
