Aqui, vou utilizar o arquivo de votação das eleições de 2024 de Roraima para trabalhar com o group_by

```python
import pandas as pd
# Importação do arquivo
dados_eleitorais = pd.read_csv("/content/votacao_candidato_munzona_2024_RR.csv", encoding="latin-1", sep=";")

print(dados_eleitorais)
```

```python
# Aqui, crio um dataframe com base nos dados eleitorais importados
dataframe_dados_eleitorais = pd.DataFrame(dados_eleitorais)

# Aqui vai o primeiro group_by: ele serve para agrupar o número do partido
# e o número de votos nominais por cada partido, ou seja,
# traz a soma da contagem de votos por part.
group_numVotos_numPart = dataframe_dados_eleitorais['QT_VOTOS_NOMINAIS'].groupby(dataframe_dados_eleitorais['NR_PARTIDO'])

# No caso, utilizei a função sum() para trazer a soma de todos os votos que
# cada partido recebeu
print(group_numVotos_numPart.sum())
```

```python
# Aqui, crio um dataframe com base nos dados eleitorais importados
dataframe_dados_eleitorais = pd.DataFrame(dados_eleitorais)

# Aqui vai o segundo group_by: ele serve para agrupar o número do partido,
# o nome do partido e o número de votos nominais por cada partido, ou seja,
# traz a soma da contagem de votos por part.
group_numVotos_numPart = dataframe_dados_eleitorais['QT_VOTOS_NOMINAIS'].groupby([dataframe_dados_eleitorais['NR_PARTIDO'], dataframe_dados_eleitorais['NM_PARTIDO']])
print(group_numVotos_numPart.sum())
```

```python
dataframe_dados_eleitorais = pd.DataFrame(dados_eleitorais)

# Aqui, ele faz a contagem de quantos vereadores (código 13) e
# quantos prefeitos estiveram na disputa
print(dataframe_dados_eleitorais["CD_CARGO"].value_counts())
```

```python
dataframe_dados_eleitorais = pd.DataFrame(dados_eleitorais)

# Aqui, eu agrupo o número de votos nominais com o código do cargo.
group_numVotos_numPart = dataframe_dados_eleitorais['QT_VOTOS_NOMINAIS'].groupby(dataframe_dados_eleitorais['CD_CARGO'])

# Assim, sei qual foi o número médio de votos de cada prefeito e
# de cada vereador através do mean()
print(group_numVotos_numPart.mean())

# Já aqui ele retorna o número max de votos que um vereador ou prefeito recebeu
print(group_numVotos_numPart.max())

# Aqui é o mínimo
print(group_numVotos_numPart.min())
```
