# Series

```python
import pandas as pd

# Cria um dicionário com as chaves e os valores
dic_alunos = {'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'}
# Transforma o dicionário em uma séries / series, convertendo as chaves em índices
# e os valores em rótulos
alunos = pd.Series(dic_alunos)

# Printa tanto o índice quanto os valores
print(alunos)
# Printa o rótulo no índice "A1"
print(alunos.A1)
# Printa todos os índices da série
print(alunos.index)
# Printa todos os valores da série
print(alunos.values)
# Printa o rótulo na posição zero
print(alunos.values[0])
# Printa o valor na posição zero
print(alunos.index[0])
# Printa o tamanho da série
print(alunos.size)
```

```python
# Indexação de series
dic_alunos = {'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'}

# Padrão
print(alunos.values[0])
print(alunos.A2)
print(alunos.values[-2])

# Por slicing
print(alunos.values[0:])
print(alunos.values[0:2])
print(alunos.values[-3:])
```

```python
import pandas as pd
# Indexação Booleana
notas = pd.Series([7.6, 5.0, 8.5, 9.5, 6.4])
alunos = pd.Series({'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'})

# obtém os índices dos alunos aprovados
idx_aprovados = list(notas[notas >= 7].index)

print(idx_aprovados)
# Para trazer por posição específica (nesse caso 0, 2, 3) deve-se usar o .iloc
print(alunos.iloc[idx_aprovados])
```

```python
titulos_copadomundo = pd.Series({
    'BR': 5, # Brasil
    'DE': 4, # Alemanha
    'IT': 4, # Itália
    'AR': 3, # Argentina
    'FR': 2, # França
    'UY': 2, # Uruguai
    'EN': 1, # Inglaterra
    'ES': 1  # Espanha
})

maiores_ganhadores = list(titulos_copadomundo[titulos_copadomundo >= 3].index)

print(titulos_copadomundo[maiores_ganhadores])
```

```python
# Busca
alunos = pd.Series({'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'})

# Nesses casos, a pergunta dos ifs vai sempre retornar True ou False
if "Dayse" in alunos.values:
    print("Sim, tem")
if "Leo" not in alunos.values:
    print("Não tem")

if "A1" in alunos.index:
    print("Sim, tem")
if "A6" not in alunos.index:
    print("Não tem")

# Para verificar se uma lista pertence, usar o método isin()
```

```python
# Modificação
alunos = pd.Series({'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'})

# Adicionar
alunos["A6"] = "Fulano"
print(alunos)

# Alterar
alunos["A3"] = "Beltrano"
print(alunos)

# Remover
alunos.drop("A4")
print(alunos)
```

```python
# Iteração
alunos = pd.Series({'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'})

for _ in alunos:
    print(_)

for _ in alunos.index:
    print(_)
```

```python
# Aritmética
a = pd.Series({
    'BR': 5, # Brasil
    'DE': 4, # Alemanha
    'IT': 4, # Itália
    'AR': 3, # Argentina
    'FR': 2, # França
    'UY': 2, # Uruguai
    'EN': 1, # Inglaterra
    'ES': 1  # Espanha
})
b = pd.Series({
    'BR': 5, # Brasil
    'DE': 4, # Alemanha
    'IT': 4, # Itália
    'AR': 3, # Argentina
    'FR': 2, # França
    'UY': 2, # Uruguai
    'EN': 1, # Inglaterra
    'ES': 1  # Espanha
})

print(a * b)
print(a / b)
print(a + b)
print(a - b)
```
