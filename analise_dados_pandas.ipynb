# Series

```python
import pandas as pd

# Cria um dicionário com as chaves e os valores
dic_alunos = {'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'}
# Transforma o dicionário em uma séries / series, convertendo as chaves em índices
# e os valores em rótulos
alunos = pd.Series(dic_alunos)

# Printa tanto o índice quanto os valores
print(alunos)
# Printa o rótulo no índice "A1"
print(alunos.A1)
# Printa todos os índices da série
print(alunos.index)
# Printa todos os valores da série
print(alunos.values)
# Printa o rótulo na posição zero
print(alunos.values[0])
# Printa o valor na posição zero
print(alunos.index[0])
# Printa o tamanho da série
print(alunos.size)
```

```python
import pandas as pd
# Indexação Booleana
notas = pd.Series([7.6, 5.0, 8.5, 9.5, 6.4])
alunos = pd.Series({'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'})

# obtém os índices dos alunos aprovados
idx_aprovados = list(notas[notas >= 7].index)

print(idx_aprovados)
# Para trazer por posição específica (nesse caso 0, 2, 3) deve-se usar o .iloc
print(alunos.iloc[idx_aprovados])
```

### Operações básicas

```python
# Criação de um dataframe
import pandas as pd
# Aqui, os dados criam as colunas nome, continente, extensão e corVerde.
# Assim, as colunas são um dicionário, e cada coluna é uma series
dados = {'nome': ['Argentina','Brasil','França','Itália',
'Reino Unido'],
'continente': ['América','América','Europa','Europa',
'Europa'],
'extensao': [2780,8511,644,301,244],
'corVerde': [0,1,0,1,0]
}
# Aqui, se faz que o índice da Series dados é a sigla de cada país
siglas = ['AR','BR','FR','IT','UK']
paises = pd.DataFrame(dados,index=siglas)

print(paises)
# Retorna os valores das colunas
print(paises.columns.values)
# Retorna os valores dos índices
print(paises.index.values)
print("Seu dataframe tem ", paises.shape[0] , "linhas")
print("Seu dataframe tem ", paises.shape[1] , "colunas")
```

```python
# Indexação - linhas
import pandas as pd

# Printa o primeiro conjunto de valor (ou seja, a primeira linha)
print(paises.values[0])
# Printa da primeira até a última linha
print(paises.values[0:])
# Printa a primeira até a terceira linha
print(paises.values[0:3])
# Printa a primeira até a útlima linha, pulando de 2 em 2
print(paises.values[0::2])
# Printa da última até a primeira linha
print(paises.values[-1:])
```

```python
# Indexação - linhas + colunas

# Printa a informação da linha 1 e coluna 3
print(paises.values[0][2]) #Ou
print(paises.values[0,2])
# Printa as informações vinculadas a determinado índice/rótulo
# O loc() e o at() servem para procurar através de rótulo
print(paises.loc["BR"])
print(paises.loc["BR", "continente"])
# Printa as informações vinculadas a determinado índice/rótulo
# O iloc() e o iat() servem para procurar através de rótulo através de inteiros
print(paises.iloc[1])

# Printa os valores de uma coluna
print(paises["nome"])
print(paises["nome"].values)
```

```python
# Indexação - fatiamento

# Por fatimento, primeiro vem o intervalo das linhas e depois o das colunas
print(paises.iloc[0:3, 0:])
print(paises.loc["AR":, "nome":"extensao"])
print(paises.iloc[-2:, :])
```

```python
#Checagem

if "BR" in paises.index or paises.values:
    print("Tem o índice BR")
if "Europa" in paises.values:
    print("Tem Europa nos valores")
if "nome" in paises.columns:
    print("Tem a coluna nome")

# Verifica se tal valor faz parte uma coluna
if "França" in paises["nome"].values:
    print("Tem o país frança na coluna nome")

# Verifica de acordo com determinado índice
if "França" in paises.loc["AR"]:
    print("Tem o país frança no índice AR")
else:
    print("Não tem o país frança no índice AR")
```
