# Series

```python
import pandas as pd

# Cria um dicionário com as chaves e os valores
dic_alunos = {'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'}
# Transforma o dicionário em uma séries / series, convertendo as chaves em índices
# e os valores em rótulos
alunos = pd.Series(dic_alunos)

# Printa tanto o índice quanto os valores
print(alunos)
# Printa o rótulo no índice "A1"
print(alunos.A1)
# Printa todos os índices da série
print(alunos.index)
# Printa todos os valores da série
print(alunos.values)
# Printa o rótulo na posição zero
print(alunos.values[0])
# Printa o valor na posição zero
print(alunos.index[0])
# Printa o tamanho da série
print(alunos.size)
```

```python
import pandas as pd
# Indexação Booleana
notas = pd.Series([7.6, 5.0, 8.5, 9.5, 6.4])
alunos = pd.Series({'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'})

# obtém os índices dos alunos aprovados
idx_aprovados = list(notas[notas >= 7].index)

print(idx_aprovados)
# Para trazer por posição específica (nesse caso 0, 2, 3) deve-se usar o .iloc
print(alunos.iloc[idx_aprovados])
```

```python
# Busca
alunos = pd.Series({'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'})

if "Dayse" in alunos.values:
    print("Sim, tem")
if "Leo" not in alunos.values:
    print("Não tem")

if "A1" in alunos.index:
    print("Sim, tem")
if "A6" not in alunos.index:
    print("Não tem")

# Para verificar se uma lista pertence, usar o método isin()
```

```python
# Modificação
alunos = pd.Series({'A1':'Bob','A2':'Dayse','A3':'Bill',
'A4':'Cris','A5':'Jimi'})

# Adicionar
alunos["A6"] = "Fulano"
print(alunos)

# Alterar
alunos["A3"] = "Beltrano"
print(alunos)

# Remover
alunos.drop("A4")
print(alunos)
```
