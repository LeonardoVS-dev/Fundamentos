```python
# Aplicação de map()
def main():
    # Defino o contador i como 1
    i = 1

    # Definição do dataframe notas
    notas = pd.DataFrame({"A1":[9.8, 7.2, 8.0],
    "A2": [5.3, 4.0, 3.5],
    "A3": [5.5, 8.1, 7.2],
    "A4": [7.0, 7.5, 6.5]},
    index=["P1","P2","P3"])

    # Aqui, utilizando o contador i, ele irá iterar enquanto i for menor ou
    # igual ao número de colunas
    while i <= len(notas.columns):
        # Defino a varável number como a string de i
        number = str(i)
        # Defino que a variável resultado equivale às notas de cada aluno,
        # mapeadas pela função verificacao_notas
        resultado = notas['A'+number].map(verificacao_notas)
        print(resultado)
        # Aumento o valor de i em 1
        i += 1
        print("\n")

# Verifico as notas de cada aluno, com base na coluna do DF notas, de
# acordo com a prova
def verificacao_notas(input):
    if (input >= 6):
        return "Resultado: Aprovado"
    elif (input >= 5) & (input < 6):
        return "Resultado: Recuperação"
    else:
        return "Resultado: Reprovado"

main()
```
