Aplicação das funções apply, map, transform, applymap

Função, aplicação e tipo de retorno de apply, map, transform e applymap.


| Função | Aplicação | Retorno
| ------ | --------- | -------
| Apply  | Dataframes e series | Igual ao que foi passado originalmente
| Map | Apenas Series | pandas.Categorical ou pandas.Index
| Transform | Dataframes e series | Igual ao que foi passado originalmente
| Applymap | descontinuado | descontinuado


O transform e o apply são muito parecidos, mas há algumas diferanças:
- o apply pode operar em Dataframes ou Series inteiros, enquanto o transform opera em Series individuais;
- saída: o apply pode retornar várias estruturas de dados, enquanto o transform deve retornar uma Series com o mesmo tamanho da de entrada;
- apply pode mudar o índice, enquanto o transform preserva o índice;

Fonte: https://medium.com/@stacymacbrains/difference-between-apply-and-transform-in-pandas-a975b018e862

```python
# Aplicação de map()
def main():
    # Defino o contador i como 1
    i = 1

    # Definição do dataframe notas
    notas = pd.DataFrame({"A1":[9.8, 7.2, 8.0],
    "A2": [5.3, 4.0, 3.5],
    "A3": [5.5, 8.1, 7.2],
    "A4": [7.0, 7.5, 6.5]},
    index=["P1","P2","P3"])

    # Aqui, utilizando o contador i, ele irá iterar enquanto i for menor ou
    # igual ao número de colunas
    while i <= len(notas.columns):
        # Defino a varável number como a string de i
        number = str(i)
        # Defino que a variável resultado equivale às notas de cada aluno,
        # mapeadas pela função verificacao_notas. Apesar do map
        # servir apenas para series, deve-se entender que cada coluna
        # num dataframe é uma series
        resultado = notas['A'+number].map(verificacao_notas)
        print(resultado)
        # Aumento o valor de i em 1
        i += 1
        print("\n")

# Verifico as notas de cada aluno, com base na coluna do DF notas, de
# acordo com a prova
def verificacao_notas(input):
    if (input >= 6):
        return "Resultado: Aprovado"
    elif (input >= 5) & (input < 6):
        return "Resultado: Recuperação"
    else:
        return "Resultado: Reprovado"

main()
```

```python
# transform()
import pandas as pd

def main():
    notas = pd.DataFrame({"A1":[9.8, 7.2, 8.0],
    "A2": [5.3, 4.0, 3.5],
    "A3": [5.5, 8.1, 7.2],
    "A4": [7.0, 7.5, 6.5]},
    index=["P1","P2","P3"])

    resultado = notas.transform(ponto_extra, axis=0)
    print(resultado)

def ponto_extra(input):
    if (input <= 9).all():
        return input + 1
    else:
        return input

main()
```
